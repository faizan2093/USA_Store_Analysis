```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load Excel File in Excel format.
dataset = pd.read_excel("Superstore_USA.xlsx")
```

```python
#Show information in Dataset Column , Values and datatype.
dataset.info()
```

```python
#Show data in dataset.
dataset.head(2)
```

```python
#Check Data Shape in Dataset
dataset.shape
```

```python
#Check is Null Value in data.
dataset.isnull().sum()
```

```python
#Fill mean Values in Black Rows.
dataset['Product Base Margin'] = dataset['Product Base Margin'].fillna(dataset['Product Base Margin'].mean())
```

Count of Order Priority

```python
#Graphical Table 
plt.figure(figsize=(5,4))
sns.countplot(x="Order Priority",data=dataset)
plt.title("Count of Order Priority")
plt.savefig("Count of Order Priority.jpg")
plt.show()
```

Ship Mode

```python
dataset['Ship Mode'].value_counts()
```

```python
x = dataset['Ship Mode'].value_counts().index
y = dataset['Ship Mode'].value_counts().values
```

```python
plt.figure(figsize=(5,4))
plt.pie(y,labels=x, startangle=60, autopct="%0.2f%%")
plt.legend(loc=3)
plt.savefig("Ship Mode.jpg")
plt.show()
```
